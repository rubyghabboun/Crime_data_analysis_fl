Total arrests and crime by county.

```python
import pandas as pd
```

```python
df = pd.read_excel("Resources/Total_Arrests_by_County_2020.xlsx")
```

```python
df.head()
```

```python
df.columns
```

```python
columns_to_keep = ['County', 'Year', 'Total Arrests', 'Total Adult Arrests', 'Total Juvenile Arrests',
                   'Murder', 'Rape', 'Robbery', 'Burglary', 'Larceny', 'Motor Vehicle Theft',
                   'Stolen Property','Kidnap/\nAbduction', 'Drug Arrest', 'Aggravated Assault']

# Select only the desired columns
df_filtered = df[columns_to_keep]
df_filtered.head()
```

```python
columns_to_convert = ['Year', 'Total Arrests', 'Total Adult Arrests', 'Total Juvenile Arrests',
                      'Murder', 'Rape', 'Robbery', 'Burglary', 'Larceny', 'Motor Vehicle Theft',
                      'Stolen Property', 'Kidnap/\nAbduction', 'Drug Arrest', 'Aggravated Assault']

df_filtered[columns_to_convert] = df_filtered[columns_to_convert].fillna(0).astype(int)

# View the converted columns
df_filtered
```

```python
#merge Robbery,Burglary,Larceny,Motor Vehicle Theft, Stolen Property to one column ( thift)
df_filtered['Theft'] = df_filtered[['Robbery', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Stolen Property']].sum(axis=1)

# Drop the individual columns (Robbery, Burglary, Larceny, Motor Vehicle Theft, Stolen Property)
df_filtered = df_filtered.drop(['Robbery', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Stolen Property'], axis=1)

# View the modified DataFrame
df_filtered.head()
```

```python
df_filtered.tail()
```

```python
df_filtered = df_filtered.drop(67)

# Reset the index after dropping the row
df_filtered = df_filtered.reset_index(drop=True)
```

```python
# Drop row 68 florida totals
df_filtered = df_filtered.drop(index=67)
```

```python
df_filtered = df_filtered.rename(columns={'Kidnap/\nAbduction': 'Kidnap and Abduction'})
```

```python
df_filtered.tail()
```

```python
df_filtered.to_csv('filtered_data2.csv', index=False)
df_filtered.to_excel('filtered_data2.xlsx', index=False)
```
