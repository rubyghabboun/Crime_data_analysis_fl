```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_names= ['FIBRS_2021.csv', 'FIBRS_2022 .csv', 'FIBRS_2023_Q2 .csv']
dfs = [pd.read_csv(f'data/{file}', header=[0, 1]) for file in file_names]
df = pd.concat(dfs, ignore_index=True)
```

```python
df.head(5)
```

```python
plt.figure(figsize=(12, 8))

# Iterate over unique counties
for county in df['County'].unique():
    county_data = df[df['County'] == county]

    # Plot line for each agency
    for index, row in county_data.iterrows():
        crime_categories = ['Violent Crime', 'Property Crime', 'Other Crime']
        annual_rates = [row[f'{category} Annual Rate'] for category in crime_categories]

        plt.plot(crime_categories, annual_rates, label=f"{row['Agency Name']} - {row['County']}")

# Customize the chart
plt.title('Crime Rates by Agency and County (2022)')
plt.xlabel('Crime Category')
plt.ylabel('Annual Rate')
plt.legend()
plt.grid(True)
plt.show()
```
