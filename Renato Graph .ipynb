```python
ctual column names
data_filtered = df_2021[df_2021['County'].isin(['Miami-Dade', 'Palm Beach'])][['County', 'Violent Crime Year to Date', 'Property Crime Year to Date', 'Other Crime Year to Date', 'Agency Name']]

# Set 'County' and 'Agency Name' as the index
data_filtered.set_index(['County', 'Agency Name'], inplace=True)

# Convert columns to numeric
numeric_cols = ['Violent Crime Year to Date', 'Property Crime Year to Date', 'Other Crime Year to Date']
data_filtered[numeric_cols] = data_filtered[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Plotting a stacked bar chart
ax = data_filtered.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Crime Distribution in 2021 for Miami-Dade and Palm Beach Counties')
plt.xlabel('(County, Agency Name)')
plt.ylabel('Crime Count')
plt.legend(title='Crime Type')
plt.show()
```
